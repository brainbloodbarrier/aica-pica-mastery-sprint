```python
# Interactive question function
def display_question(q):
    """Display a single question with interactive answer selection"""
    display(Markdown(f"""
### Question {q['id']}
    
**Category**: {q['category']} | **Difficulty**: {q['difficulty']}
    
---
    
{q['question']}
    """))
    
    radio = RadioButtons(
        options=q['options'],
        layout=Layout(width='100%')
    )
    
    submit_btn = Button(description='Submit Answer', button_style='primary')
    output = Output()
    
    def on_submit(btn):
        with output:
            clear_output()
            is_correct = radio.value == q['correct']
            
            # Update score
            score['total'] += 1
            if is_correct:
                score['correct'] += 1
            
            # Track by category
            if q['category'] not in score['by_category']:
                score['by_category'][q['category']] = {'correct': 0, 'total': 0}
            score['by_category'][q['category']]['total'] += 1
            if is_correct:
                score['by_category'][q['category']]['correct'] += 1
            
            if is_correct:
                print("‚úÖ CORRECT!\n")
            else:
                print(f"‚ùå INCORRECT.\n")
                print(f"The correct answer is: {q['correct']}\n")
            
            print(f"üìù Explanation:\n{q['explanation']}")
            print("\n" + "="*80)
            print(f"Your Score: {score['correct']}/{score['total']} ({(score['correct']/score['total']*100):.1f}%)")
            print("="*80)
    
    submit_btn.on_click(on_submit)
    
    display(VBox([radio, submit_btn, output]))

print("‚úÖ Question display function ready")
```

### Key Clinical Syndromes

**AICA Occlusion (Lateral Pontine Syndrome)**:
- CN VII palsy (ipsilateral)
- CN VIII dysfunction: hearing loss, vertigo (ipsilateral)
- Facial sensory loss (ipsilateral, spinal trigeminal tract)
- Horner syndrome (ipsilateral)
- Cerebellar ataxia (ipsilateral)
- Incomplete contralateral body pain/temperature loss

**PICA Occlusion (Wallenberg/Lateral Medullary Syndrome)**:
- Dysphagia, hoarseness (nucleus ambiguus)
- Facial pain/temperature loss (ipsilateral, spinal trigeminal)
- Body pain/temperature loss (contralateral, spinothalamic)
- Horner syndrome (ipsilateral)
- Vertigo, nystagmus (vestibular nuclei)
- Ipsilateral ataxia
- **NO hemiparesis** (pyramidal tract spared)

### Surgical Pearls Summary

**Universal Principles**:
1. Arachnoid plane dissection protects perforators
2. Perforators are end-arteries - never sacrifice
3. Cortical branches have collaterals - can sacrifice if needed
4. Preoperative angiography mandatory for complex cases
5. Know patient's specific anatomy - variability is extreme

**AICA-Specific**:
- Identify premeatal segment early in retrosigmoid approach
- Check for meatal loop before IAC drilling (50% enter canal)
- Preserve labyrinthine artery for hearing
- Expect duplication (26%)

**PICA-Specific**:
- Check caudal loop position relative to foramen magnum
- Cranial loop defines lateral boundary in telovelar approach
- Most perforators from tonsillomedullary segment (3.3 avg)
- 16% absent - check for compensation
- Recognize extradural origin if present (rare)
